```python
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd 
#read data from wikipedia website and get data from table in url
html = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M') 
b = BeautifulSoup(html.text, 'lxml') 
my_table = b.find('table',{'class':'wikitable sortable'})
my_table 
#pull data from table in Url
A=[]
B=[]
C=[]
D=[]
for row in my_table.findAll("tr"):
    cells = row.findAll('td')
    states = row.findAll('th')
    if len(cells)>=2:
        A.append(cells[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
#pulling selected column from table
df=pd.DataFrame(A, columns=['Postcode'])
df['Borough']=C
df['Neighbourhood']=D
df = df[df.Borough !='Not assigned'] 
df.columns= ['Postcode', 'Borough', 'Neighbourhood']
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index() 
df.shape
```

```python

filtered_columns = ['name', 'categories'] + [col for col in df.columns if col.startswith('location.')] + ['id']

type(filtered_columns)
filtered_columns

df_filtered = df.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
df_filtered['categories'] = df_filtered.apply(get_category_type, axis=1)
df_filtered.columns = [column.split('.')[-1] for column in df_filtered.columns ]
df_filtered.head()
```
